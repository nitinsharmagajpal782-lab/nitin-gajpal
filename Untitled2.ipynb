```python
import numpy as np
```

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr)
```

```python
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])

print(arr2)
```

```python
total_sum = np.sum(arr2)
print("Sum =", total_sum)
```

```python
total_product = np.prod(arr2)
print("Product =", total_product)
```

```python
print(np.sum(arr2, axis=0))
```

```python
print(np.sum(arr2, axis=1))
```

```python
print(np.prod(arr2, axis=1))
```

```python
print(np.prod(arr2, axis=0))
```

```python
sum_result = 0
product_result = 1

for row in arr2:
    for num in row:
        sum_result += num
        product_result *= num

print("Sum =", sum_result)
print("Product =", product_result)
```

```python
for element in arr:
    print(element)
```

```python
for row in arr2:
    for element in row:
        print(element)
```

```python
for element in np.nditer(arr2):
    print(element)
```

```python
arr = np.array([1, 2, 3, 4, 5])

squares = []

for num in arr:
    squares.append(num ** 2)

print(squares)
```

```python
arr = np.array([1, 2, 3, 4, 5])

squares = arr ** 2

print(squares)
```
